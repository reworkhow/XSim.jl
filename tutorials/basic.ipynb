##### Modularism of XSim

````julia
for i in 1:5
    progenies = mate(sires, dams; args_mate...)
    progenies = select(progenies, 10; args_select...)
    sires, dams = progenies, progenies
end
````

### Complete code

````julia
# Load XSim
using XSim
import Random
Random.seed!(95616)

# Build genome and phenome
build_demo()

# Initialize founders
n_sires = 5
n_dams  = 3
sires   = Founders(n_sires)
dams    = Founders(n_dams)

# Define parameters
args     = Dict(# mating
                :n_per_shared => n_dams,
                :n_per_mate   => 2,
                # selection
                :h2           => [.5, .5],
                # breeding
                :n_gens       => 5,
                :n_select     => 10)

# Breeding program
sires, dams   = breed(sires, dams; args...)
````
